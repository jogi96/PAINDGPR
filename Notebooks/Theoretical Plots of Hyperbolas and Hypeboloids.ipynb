# Formale Darstellung von Hyperboloiden und Beugungshyperbeln anhand den Grundlagen


```python
#import von be√∂tigten Packages
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
```

## Darstellung einer Hyperbel

```python
#Parameters

x0, y0, z = 5, 5, 3  # Position Diffraktor
v = 1  # Geschwindigkeit

#x,y Gitter
x = np.linspace(0, 10, 100)

#Laufzeitfunktion
T = (2.0 / v) * np.sqrt((x - x0)**2 + z**2)

#Plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, T)
ax.plot(x0, (2.0 / v) * z, 'ro')
ax.set_xlabel('x')
ax.set_ylabel('Laufzeit T')
ax.set_title('Laufzeitfunktion eines Diffraktors')
ax.invert_yaxis()
plt.legend(['Laufzeit T', 'Diffraktor'])
plt.grid()
plt.show()
```

## Darstellung eines Hyperbolids in 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from matplotlib.lines import Line2D

# Parameter
x0, y0, z = 5, 5, 3 
v = 1 

# x,y-Gitter
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# Laufzeitfunktion
T = (2.0 / v) * np.sqrt((X - x0)**2 + (Y - y0)**2 + z**2)

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')

# Laufzeitfl√§che plotten
ax.plot_surface(X, Y, T, cmap='viridis', alpha=0.95, linewidth=0)

# Punktdiffraktor
ax.scatter(x0, y0, (2.0 / v) * z -0.2,
           color='yellow', s=100, edgecolor='black', zorder=10)

# Konturen
ax.contour(X, Y, T, zdir='z', offset=np.min(T), cmap='viridis', linestyles="dashed")
ax.contour(X, Y, T, zdir='x', offset=np.min(x), cmap='viridis', linestyles="dashed")
ax.contour(X, Y, T, zdir='y', offset=np.max(y), cmap='viridis', linestyles="dashed")

# Achsenbeschriftung
ax.set_xlabel('X Achse')
ax.set_ylabel('Y Achse')
ax.set_zlabel('Laufzeit',labelpad=0)
ax.set_title('Laufzeitfunktion eines Diffraktors in 3D')
ax.set(xlim=(np.min(x), np.max(x)),
       ylim=(np.min(y), np.max(y)),
       zlim=(np.min(T), np.max(T)))
ax.view_init(elev=25, azim=110)

ax.invert_zaxis()


# Farbskala f√ºr die Oberfl√§che
sm = ScalarMappable(cmap='viridis')
sm.set_array([])

# Erstelle den Legenden-Eintrag f√ºr die Farbfl√§che und den Punkt
legend_elements = [
    Line2D([0], [0], color='none', marker='s',
           markerfacecolor=plt.cm.viridis(0.5), markeredgecolor='black',
           markersize=12, label='Laufzeit T'),
    Line2D([0], [0], color='none', marker='o',
           markerfacecolor='yellow', markeredgecolor='black',
           markersize=10, label='Diffraktor')
]

ax.legend(handles=legend_elements, loc='upper left')

plt.show()
```

## 2d Schnitte

```python
#Parameters

x0, y0, z = 5, 5, 3  # Position Diffraktor
v = 1  # Geschwindigkeit

#x,y Gitter
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

#Laufzeitfunktion
T = (2.0 / v) * np.sqrt((X - x0)**2 + (Y - y0)**2 + z**2)

#Plot
fig = plt.figure(figsize=(20,20))

# x T Ebene
ax_xt = fig.add_subplot(2,3,1)
ax_xt.plot(x, T)
ax_xt.set_xlabel('X Achse')
ax_xt.set_ylabel('Laufzeit')
ax_xt.set_title("Laufzeitfunktion eines Diffraktors ind 2D in X-T Ebene")

# y T Ebene
ax_yt = fig.add_subplot(2,3,2)
ax_yt.plot(y, T)
ax_yt.set_xlabel('Y Achse')
ax_yt.set_ylabel('Laufzeit')
ax_yt.set_title("Laufzeitfunktion eines Diffraktors ind 2D in Y-T Ebene")

# x y Ebene
ax_xy = fig.add_subplot(2,3,3)
ax_xy.contour(X, Y, T, cmap='viridis')
ax_xy.set_xlabel('X Achse')
ax_xy.set_ylabel('Y Achse')
ax_xy.set_title("Laufzeitfunktion eines Diffraktors ind 2D in X-Y Ebene")   

#Plot
ax_3d = fig.add_subplot(2,3,5,projection='3d')

ax_3d.plot_surface(X, Y, T, cmap='viridis')
ax_3d.contour(X, Y, T, zdir='z', offset=np.min(T), cmap='viridis', linestyles="dashed")
ax_3d.contour(X, Y, T, zdir='x', offset=np.min(x), cmap='viridis', linestyles="dashed")
ax_3d.contour(X, Y, T, zdir='y', offset=np.max(y), cmap='viridis', linestyles="dashed")
ax_3d.set(xlim=(np.min(x),np.max(x) ), ylim=(np.min(y),np.max(y)), zlim=(np.min(T),np.max(T)))
ax_3d.set_xlabel('X Achse')
ax_3d.set_ylabel('Y Achse')
ax_3d.set_zlabel("Laufzeit")
ax_3d.set_title("Laufzeitfunktion eines Diffraktors ind 3D")
plt.show()
```

## Simulation der einzelnen Schnitte 

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from IPython.display import HTML
from matplotlib.animation import PillowWriter


# Parameters
x0, y0, z = 5, 5, 3  # Position Diffraktor
v = 5  # Geschwindigkeit

# x, y Gitter
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

T = (2.0 / v) * np.sqrt((X - x0)**2 + (Y - y0)**2 + z**2)
T_min, T_max = np.min(T), np.max(T)
times = np.linspace(T_min, T_max, 100)

fig = plt.figure(figsize=(16,10))

ax_yt = fig.add_subplot(2,3,1)
ax_xt = fig.add_subplot(2,3,2)
ax_xy = fig.add_subplot(2,3,3)
ax_3dy = fig.add_subplot(2,3,4,projection='3d')
ax_3dx = fig.add_subplot(2,3,5,projection='3d')
ax_3dxy = fig.add_subplot(2,3,6,projection='3d')


# Animation: x-Schnitt durch die y-T-Ebene
yt, = ax_yt.plot([], [],'ro', lw=2)
ax_yt.set_xlim(np.min(y), np.max(y))
ax_yt.set_ylim(0, np.max((2.0 / v) * np.sqrt((x - x0)**2 + (y - y0)**2 + z**2)))
ax_yt.set_xlabel('y')
ax_yt.set_ylabel('Laufzeit T')
ax_yt.set_title('Schnitt durch y-T Ebene f√ºr verschiedene x')

#Animation x und t Schnitt
xt, = ax_xt.plot([], [],'ro', lw=2)
ax_xt.set_xlim(np.min(x), np.max(x))
ax_xt.set_ylim(0,np.max((2.0 / v) * np.sqrt((x - x0)**2 + (y - y0)**2 + z**2)))
ax_xt.set_xlabel('x')
ax_xt.set_ylabel('Laufzeit T')
ax_xt.set_title('Schnitt durch x-T Ebene f√ºr verschiedene x')

#animation xy Ebene
xy, = ax_xy.plot([], [], 'ro', lw=2)
ax_xy.set_xlim(0, 10)
ax_xy.set_ylim(0, 10)
ax_xy.set_xlabel('x')
ax_xy.set_ylabel('y')
ax_xy.set_title('Position in der xy-Ebene')


#Animation 3d f√ºr yt Schnitt
surface = ax_3dy.plot_surface(X,Y,T, cmap='viridis') #Hyperboloid

x_plane = np.full_like(Y,x[0])
T_planeyt = (2.0 / v) * np.sqrt((x_plane - x0)**2 + (Y - y0)**2 + z**2)
planeyt = ax_3dy.plot_surface(x_plane,Y,T_planeyt,color='red',alpha=0.4)
ax_3dy.set(xlim=(0,10 ), ylim=(0,10), zlim=(np.min(T),np.max(T)))
ax_3dy.set_xlabel('X Achse')
ax_3dy.set_ylabel('Y Achse')
ax_3dy.set_zlabel("Laufzeit")
ax_3dy.set_title('Ansicht 3d wenn man sich in richtung x bewegt')


#Animation 3d f√ºr xt Schnitt
surface = ax_3dx.plot_surface(X,Y,T, cmap='viridis') #Hyperboloid

y_plane = np.full_like(X,y[0])
T_planext = (2.0 / v) * np.sqrt((X - x0)**2 + (y_plane - y0)**2 + z**2)
planext = ax_3dx.plot_surface(X,y_plane,T_planext,color='red',alpha=0.4)
ax_3dx.set(xlim=(0,10 ), ylim=(0,10), zlim=(np.min(T),np.max(T)))
ax_3dx.set_xlabel('X Achse')
ax_3dx.set_ylabel('Y Achse')
ax_3dx.set_zlabel("Laufzeit")
ax_3dx.set_title('Ansicht 3d wenn man sich in richtung y bewegt')

#Animation 3d f√ºr xy Schnitt
surface = ax_3dxy.plot_surface(X,Y,T, cmap='viridis') #Hyperboloid

t_val = T_min
r = np.sqrt((v * t_val / 2)**2 - z**2) if (v * t_val / 2)**2 > z**2 else 0
theta = np.linspace(0, 2*np.pi, 400)
x_circle = x0 + r * np.cos(theta)
y_circle = y0 + r * np.sin(theta)
z_circle = np.full_like(theta, t_val)
circle, = ax.plot(x_circle, y_circle, z_circle, 'r', lw=2)
ax_3dxy.set_xlabel('X Achse')
ax_3dxy.set_ylabel('Y Achse')
ax_3dxy.set_zlabel("Laufzeit")
ax_3dxy.set_title('Ansicht 3d wenn man sich in richtung T bewegt')

def update(frame):
    
    x_val = float(x[frame])
    y_val = float(y[frame])
    t_val = float(T[frame,frame])
    #update yt plot
    T_sliceyt = (2.0 / v) * np.sqrt((x_val - x0)**2 + (y - y0)**2 + z**2)
    yt.set_data(y, T_sliceyt)
    ax_yt.set_title(f'Schnitt bei x={x_val:.2f}')

    #updatext plot
    T_slicext = (2.0 / v) * np.sqrt((x - x0)**2 + (y_val - y0)**2 + z**2)
    xt.set_data(x, T_slicext)
    ax_xt.set_title(f'Schnitt bei y={y_val:.2f}')
    
    
    #update Plane in 3d yt
    global planext
    planext.remove()
    x_plane = np.full_like(Y, x_val)
    T_planeyt = (2.0 / v) * np.sqrt((x_plane - x0)**2 + (Y - y0)**2 + z**2)
    planext = ax_3dy.plot_surface(x_plane, Y, T_planeyt, color='red', alpha=0.4, edgecolor='k')
    ax_3dy.set_title(f"x-Ebene bei x={x_val:.2f}")

    #update Plane in 3d yt
    global planeyt
    planeyt.remove()
    y_plane = np.full_like(X,y_val)
    T_planext = (2.0 / v) * np.sqrt((X - x0)**2 + (y_plane - y0)**2 + z**2)
    planeyt = ax_3dx.plot_surface(X, y_plane, T_planext, color='red', alpha=0.4, edgecolor='k')
    ax_3dx.set_title(f"y-Ebene bei y={y_val:.2f}")

    #update Plane in 3d xy
    global  circle
    circle.remove()
    
    
    
    t_val = (2.0 / v) * np.sqrt((x_val - x0)**2 + (y_val - y0)**2 + z**2)
    r_sq = (v * t_val / 2)**2 - z**2
    if r_sq <= 0:
        r = 0
    else:
        r = np.sqrt(r_sq)
    

    theta = np.linspace(0, 2*np.pi, 400)
    x_circle = x0 + r * np.cos(theta)
    y_circle = y0 + r * np.sin(theta)
    z_circle = np.full_like(theta, t_val)
    
    #Update xy plot
    xy.set_data(x_circle, y_circle)
    ax_xy.set_title(f'Laufzeit bei t= ({t_val:.2f})')
    #update Plane in 3d xy
    circle, = ax_3dxy.plot(x_circle, y_circle, z_circle, 'r', lw=2)
    ax_3dxy.set_title(f"t-Ebene bei t={t_val:.2f}")





   
   


    return yt, xt, planext, planeyt, circle

ani = animation.FuncAnimation(fig, update, frames=len(x), interval=50)
ani.save("anim.gif", writer=PillowWriter(fps=20))
```
